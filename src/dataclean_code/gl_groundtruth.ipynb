```python
import ast


# 设置关键词（你可以换成任何想找的关键词）
keyword = "store"

# 解析 category 字段
def parse_category(cat):
    if isinstance(cat, list):
        return cat
    try:
        return ast.literal_eval(cat)
    except:
        return []

df_meta['category_list'] = df_meta['category'].apply(parse_category)

# 过滤出包含关键词（不区分大小写）的商家
df_meta_filtered = df_meta[df_meta['category_list'].apply(
    lambda cat_list: any(keyword.lower() in c.lower() for c in cat_list)
)].copy()

# 重命名 name -> meta_name，避免冲突
df_meta_filtered.rename(columns={"name": "meta_name"}, inplace=True)

# 合并 review 和 meta 数据
df_merged = pd.merge(df_review, df_meta_filtered[['gmap_id', 'meta_name']], on='gmap_id', how='inner')

# 计算每个 meta_name 的平均评分
df_avg_rating = df_merged.groupby('meta_name')['rating'].mean().reset_index()

# 选出评分最高的前 10 个店铺
df_top10 = df_avg_rating.sort_values(by='rating', ascending=False)

df_top10
```

```python
import ast
import pandas as pd

# 确保 category 是列表形式
def parse_category(cat):
    if isinstance(cat, list):
        return cat
    try:
        return ast.literal_eval(cat)
    except:
        return []

# 解析 category 字段
df_meta['category_list'] = df_meta['category'].apply(parse_category)

# 拆解多类别为多行
df_exploded = df_meta[['gmap_id', 'category_list']].explode('category_list').dropna()

# 合并评分信息
df_merged = pd.merge(df_review[['gmap_id', 'rating']], df_exploded, on='gmap_id', how='inner')

# 统计类别维度的指标：评论数量、平均评分、商家数量
df_stats = df_merged.groupby('category_list').agg(
    count=('rating', 'size'),                         # 总评论数
    avg_rating=('rating', 'mean'),                    # 平均评分
    business_count=('gmap_id', pd.Series.nunique)     # 不同商家数量
).reset_index()

# 取评论最多的前 10 类别
df_stats_top10 = df_stats.sort_values(by='count', ascending=False).head(10)




df_stats_top10
```

```python
import pandas as pd

# Load data if needed
# df_review = pd.read_json("your_review_file.json", lines=True)

# Convert timestamp to datetime
df_review['datetime'] = pd.to_datetime(df_review['time'], unit='ms')

# Define fixed time range: Jan 1, 2018 to Jan 1, 2019
start_date = pd.Timestamp("2019-01-01")
end_date = pd.Timestamp("2020-01-01")
df_period = df_review[(df_review['datetime'] >= start_date) & (df_review['datetime'] < end_date)]

# Filter high-score reviews (rating >= 4.5)
df_highscore = df_period[df_period['rating'] >= 4.5]

# Count per business
df_highscore_count = df_highscore.groupby('gmap_id').size().reset_index(name='highscore_count')

# Sort and get top 3
df_top3 = df_highscore_count.sort_values(by='highscore_count', ascending=False).head(3)

# Optional: join with metadata to get business name
# df_meta = pd.read_json("your_meta_file.json", lines=True)
df_top3 = pd.merge(df_top3, df_meta[['gmap_id', 'name']], on='gmap_id', how='left')

# Show result
print("Top 3 businesses with the most high-score (≥4.5) reviews in 2018:")
print(df_top3.to_string(index=False))



```
