```python
# 定义亚洲地区的指数列表
asian_indices = ["HSI", "000001.SS", "399001.SZ", "N225", "KS11", "NSEI", "TWII"]

# Filter for Asian indices
df_asian = df_process[df_process["Index"].isin(asian_indices)].copy()

# Calculate intraday volatility
df_asian["volatility"] = (df_asian["High"] - df_asian["Low"]) / df_asian["Open"]

# Convert Date and filter for 2020 and later
df_asian["Date"] = pd.to_datetime(df_asian["Date"])
df_asian_recent = df_asian[df_asian["Date"] >= "2020-01-01"].copy()

# Compute average volatility since 2020
avg_vol_recent = df_asian_recent.groupby("Index")["volatility"].mean().reset_index()
avg_vol_recent = avg_vol_recent.sort_values("volatility", ascending=False).reset_index(drop=True)

avg_vol_recent
```
